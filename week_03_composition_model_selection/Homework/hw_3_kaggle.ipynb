# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

lr = LogisticRegression(max_iter=1000)

params = [
    {
        'penalty': ['l1'],
        'solver': ['liblinear', 'saga'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100]
    },
    {
        'penalty': ['l2'],
        'solver': ['lbfgs', 'newton-cg', 'sag', 'liblinear', 'saga'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100]
    },
    {
        'penalty': ['elasticnet'],
        'solver': ['saga'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'l1_ratio': [0.1, 0.5, 0.9]
    },
    {
        'penalty': ['none'],
        'solver': ['lbfgs', 'newton-cg', 'sag', 'saga']
    }
]

lr_grid = GridSearchCV(
    lr,
    params,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

lr_grid.fit(X_train, y_train)
```

```python
import catboost
from sklearn.compose import ColumnTransformer
from catboost import Pool, cv
```

```python
y=data_train.Churn
```

```python
X=data_train.drop("Churn",axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size=0.75,
                                                    random_state=42,
                                                    stratify=y)
```

```python
train_dataset = Pool(data=X_train,
                  label=y_train,
                  cat_features=cat_cols
                  )
```

```python
model = catboost.CatBoostClassifier(loss_function='Logloss',
                                    random_state=42,
                                    eval_metric='AUC:hints=skip_train~false')
grid = {
        'iterations':[100,200,300],
        'learning_rate': [0.03,0.1,0.5,0.7],
        'depth': [4, 6, 10],
        'l2_leaf_reg': [1, 3, 5, 7, 9]}

grid_search_result = model.grid_search(grid,
                                       X=train_dataset,
                                       plot=True
                                      )
```

```python
grid_search_result['params']
```

```python
best_model = catboost.CatBoostClassifier(
  loss_function='Logloss',
   bagging_temperature=1,
   random_strength=1,
   thread_count=3,
   iterations=100,
   l2_leaf_reg = 3,
   learning_rate = 0.5,
   save_snapshot=True,
   random_seed=63,
   od_type='Iter',
   od_wait=20,
   eval_metric='AUC:hints=skip_train~false',
   depth=4
)
```

```python
best_model.fit(
   X_train, y_train,
   cat_features=cat_cols
)
```

```python
print(best_model.get_best_score())
```
